## Data Preprocessing

```python
x = data.iloc[:,1:-1]
y = data.iloc[:,-1]
x
```

```python
le = LabelEncoder()
y = le.fit_transform(y)
y
```

```python
sc = StandardScaler()
x = sc.fit_transform(x)
```

```python
x_new = pd.DataFrame(x)
x_new.describe()
```

# Model training and evaluation

## Checking average accuracy score using KFold Cross Validation 

```python
model = DecisionTreeClassifier()
scores = cross_val_score(estimator=model, X=x, y=y, cv=7,)
print('The average accuracy score is: ',scores.mean() * 100)
```

## Hyper-parameter tuning using GridSearch Cross Validation


```python
param_grid = { 
    'criterion':['gini','entropy'],
    'max_depth':[4,5,6,7,8,9,10,11,12,15,20,30,40,50,70,90,120,150]
}
final_model = GridSearchCV(model,param_grid,cv = 5)
```

```python
final_model.fit(x,y)
final_model.best_params_
```

```python
clf = DecisionTreeClassifier(criterion='gini',max_depth=4)
clf.fit(x,y)
```

```python
fscores = cross_val_score(estimator=clf, X=x, y=y, cv=7,)
print('The average accuracy score is: ',fscores.mean() * 100)
```

## Plotting the tree obtained using sklearn tree module

```python
plt.figure(figsize=(25,10))
a = plot_tree(clf, 
              feature_names=['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm',], 
              class_names=data['Species'], 
              filled=True, 
              rounded=True, 
              fontsize=14)
```
